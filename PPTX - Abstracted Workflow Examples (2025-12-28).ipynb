```python
from pptx import Presentation
from pptx.dml.color import RGBColor
from pptx.util import Inches

prs = Presentation()
slide = prs.slides.add_slide(prs.slide_layouts[6])  # blank

bg = slide.background
bg.fill.background()                # ← clear inheritance
fill = bg.fill
fill.solid()
fill.fore_color.rgb = RGBColor(203, 24, 29)  # #cb181d red
fill.fore_color.transparency = 0.6  # 60% transparent

# Add some content to see through it
left = Inches(1)
top = Inches(1)
width = Inches(8)
height = Inches(4)
txBox = slide.shapes.add_textbox(left, top, width, height)
tf = txBox.text_frame
tf.text = "This text should show through semi-transparent red background"

prs.save("test_transparent_bg.pptx")
print("Saved test_transparent_bg.pptx – open in PowerPoint to check")
```
